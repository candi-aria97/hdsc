```python
import pandas as pd
```

```python
f_d_df1= f_d.iloc[0:19000].reset_index(drop=True)
f_d_df2= f_d.iloc[19000:].reset_index(drop=True)
```

```python
assert len(f_d) == (len(f_d_df1) + len(f_d_df2))
```

```python
pd.merge(f_d_df1, f_d_df2, how="inner")
```
