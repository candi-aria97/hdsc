```python
import pandas as pd
```

```python
dfc= pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/00374/energydata_complete.csv')
```

```python
dfc.describe(include='all')
```

```python
dfc
```

```python
dfc.isna().sum()
```

```python
dfc.drop(['date','lights'], inplace=True, axis=1)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
normalised_dfc = pd.DataFrame(scaler.fit_transform(dfc), columns=dfc.columns)
features_dfc = normalised_dfc.drop(columns=['Appliances', 'rv1'])
energy_target = normalised_dfc['Appliances']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features_dfc, energy_target, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linear_model = LinearRegression()
#fit the model to the training dataset
linear_model.fit(x_train, y_train)
#obtain predictions
predicted_values = linear_model.predict(x_test)
```

```python
#MAE
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, predicted_values)
round(mae, 3) 
```

```python
#R-squared
from sklearn.metrics import r2_score
r2_score = r2_score(y_test, predicted_values)
round(r2_score, 3)
```

```python
#residual sum of squares
import numpy as np
rss = np.sum(np.square(y_test - predicted_values))
round(rss, 3)
```

```python
#root mean square error
from sklearn.metrics import  mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, predicted_values))
round(rmse, 3)
```

```python
#ridge regression
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=0.5)
ridge_reg.fit(x_train, y_train)
```

```python
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=0.4)
ridge_reg.fit(x_train, y_train)
```

```python
from sklearn.metrics import  mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, predicted_values))
round(rmse, 3)
```

```python
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha=0.001)
lasso_reg.fit(x_train, y_train)
```

```python
#new lasso regression
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha=0.001)
lasso_reg.fit(x_train, y_train)
```

```python
def get_weights_dfc(model, feat, col_name):
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_dfc = pd.DataFrame(weights).reset_index()
    weights_dfc.columns = ['Features', col_name]
    weights_dfc[col_name].round(3)
    return weights_dfc
```

```python
linear_model = LinearRegression()
linear_model.fit(x_train, y_train)
linear_model_weights = get_weights_dfc(linear_model, x_train, 'Linear_Model_Weight')
```

```python
ridge_weights_dfc = get_weights_dfc(ridge_reg, x_train, 'Ridge_Weight')
```

```python
lasso_weights_dfc = get_weights_dfc(lasso_reg, x_train, 'Lasso_weight')
```

```python
final_weights = pd.merge(linear_model_weights, ridge_weights_dfc, on='Features')
final_weights = pd.merge(final_weights, lasso_weights_dfc, on='Features')
```
